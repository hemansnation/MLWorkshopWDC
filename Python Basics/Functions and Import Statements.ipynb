# Functions
A function is a block of code which only runs when it is called.

```python
def wdc():
    print("This is a workshop")
    print("Machine Learning")

wdc()
```

# Arguments
Information or values can be passed into functions as arguments.

```python
def add(x,y):
    print(x+y)

add(5,7)
```

# Return Statement
To let a function return a value, use the return statement.

```python
def maximum(x,y):
    if x>y:
        return x
    else:
        return y

print(maximum(7,4))
```

# Defalut Arguments
Static values passed as an argument.

```python
def say(msg, time=1):
    print(msg*time)

say("Hi")
say("Bye",5)
```
